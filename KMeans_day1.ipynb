```python
import numpy as np
import matplotlib.pyplot as plt

class KMeans:
  def __init__(self, k, max_iter = 500, tol = 1e-4):
    self.k = k
    self.max_iter = max_iter
    self.tol = tol

  def fit_predict(self, X):
    n_samples, n_features = X.shape
    random_indices = np.random.choice(n_samples, self.k, replace = False)
    centroids = X[random_indices]

    for i in range(self.max_iter):
      distances = np.linalg.norm(X[: , np.newaxis] - centroids, axis = 2)
      clusters = np.argmin(distances, axis = 1)

      new_centroids = np.array([X[clusters == j].mean(axis = 0) for j in range(self.k)])

      self.plot_progress(X, centroids, clusters, i )

      if np.linalg.norm(new_centroids - centroids) < self.tol:
        break
      centroids = new_centroids

    self.centroids = centroids
    self.labels_ = clusters
    return self.labels_

  def plot_progress(self, X, centroids, clusters, iter):
    plt.scatter(X[:, 0], X[:,1], c = clusters, s=30, cmap = 'viridis', marker = 'o', alpha = 0.6)
    plt.scatter(centroids[:, 0], centroids[:,1], c = 'red', s =200, marker = 'X')
    plt.title(f'Iteration {iter}')
    plt.show()


#test
if __name__ == '__main__':
  from sklearn.datasets import make_blobs
  X, y = make_blobs(n_samples = 500, centers = 5,cluster_std = 0.3, random_state = 3)
  kmeans = KMeans(k =5)
  labels = kmeans.fit_predict(X)
```
